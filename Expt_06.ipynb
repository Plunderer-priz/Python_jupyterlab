```python
class fruits:
  def __init__(self,name,color):
    self.name = name
    self.color = color
    print(f"Name of the fruit : {self.name}, Color of fruit : {self.color}")
fruit1 = fruits("cherry","red")
fruit2 = fruits("mango","golden")
fruit3 = fruits("orange","orange")
fruit4 = fruits("kiwi","brown")
fruit5 = fruits("avacado","green")
```

```python
class fruits:
  def __init__(self,name,price,qty):
    self.name = name
    self.price = price
    self.qty = qty
    print(f"Total price of {self.qty} dozen {self.name} is {self.price*self.qty}")
fruit1 = fruits("apples",300,1)
fruit2 = fruits("mango",1200,1)
fruit3 = fruits("orange",50,2)
fruit4 = fruits("kiwi",30,1)
fruit5 = fruits("avacado",800,1)
```
