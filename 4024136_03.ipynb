#Program 1 :



Create empty list, use while loop to add 5 items to the list, print the list

```python
li = []
k=0
from random import randrange as rand
items = ["Cables","LED","Switches","Socket","Plug","RJ 45 Socket","Lamp","Gland","Junction Box","Pipes"]
while len(li)<5:
  i = items[rand(0,10)]
  li.append(i)
  print(f"Item {k+1} : {li[k]}")
  k+=1
```

#Program 2 :

Create a list of 6 elements & perform the following operations:

1. Print first 2 and last 2 elements.
2. Add 2 elements in-between & 1 at the end
3. Remove element by value & by index
4. Clear the list
5. Delete the list

```python
li = [2,2,6,6,4,8]
print(f"Original List : {li}\nFirst 2 elements : {li[:2]}\nLast 2 elements : {li[-2:]}")
li.insert(round(len(li)/2),20)
li.insert(round(len(li)/2),40)
li.append(99)
print(f"Adding 2 elemets in-between & 1 at end ==> {li}")
li.remove(2)
li.pop()
print(f"Remove element by value 2 & last index ==> {li}")
li.clear()
print(f"Clear the list ==> {li}")
del li
print("List deleted!")
```

#Program 3:

Repeat Program 2 using tuple

```python
tup = ("Bread","Monkey","Car",12,50,"Mansion")
li = list(tup)
print(f"Original List : {li}\nFirst 2 elements : {li[:2]}\nLast 2 elements : {li[-2:]}")
li.insert(round(len(li)/2),"Kingdom")
li.insert(round(len(li)/2),"Bull")
li.append("Last Word")
print(f"Adding 2 elemets in-between & 1 at end ==> {li}")
li.remove("Monkey")
li.pop()
print(f"Remove element by value (Monkey) & last index (last word) ==> {li}")
li.clear()
print(f"Clear the list ==> {li}")
del li
print("List deleted!")
```
