#**Program 4**: Number Guessing Game


````python
import random
t = 0
num = random.randint(1,100)
while t<5 :
    inpnum = int(input("\nEnter a number from 1-100: "))
    if t == 4:
      print("Good game! You have failed, please Restart the game")
      break
    else :
      if inpnum == num :
        print ("You guessed it correct! GGS")
        break
      else :
        print("Try again!")
        t += 1
      if inpnum > num:
        print("Hint : Try smaller number\n","-"*60)
      else:
        print("Hint : Try larger number\n","-"*60)
print("The Correct number was : ",num)
````
