#Program 1

Display the conditions of chemical reaction based on inputs from Temperature & Pressure Sensors.

1. If both are beyond threshold then shutdown the unit.

2. If either 1 exceeds then take action, otherwise healthy condition

```python
t = float(input("Enter the sensed Temperature (kelvin): "))
p = float(input("Enter the sensed Pressure (bar): "))
t_threshold = 1800+273
p_threshold = 80

print("Shutdown the unit!") if t>t_threshold and p>p_threshold else print("Take Action!") if (t>t_threshold or p>p_threshold) else print("Healthy Condition")
```

#Program 2

Count numbers divisible by 3 / 5 / 7 in range 1-200

[using for & while loop]

```python
count = 0
for i in range (201):
  if (i%3==0 or i%5==0 or i%7==0):
    count += 1

print(count)
```

```python
count = 0
i = 1
while i<=200:
  if (i%3==0 or i%5==0 or i%7==0):
    count += 1
  i += 1
print(count)
```

#Program 3

Handle the following errors using try - except, while performing dision of two integers

1. 0 Division Error
2. Type Error
3. Value Error

```python
try:
  x = int(input("Enter num1 : "))
  y = int(input("Enter num2 : "))
  print(f"The division operation is {x/y}\nThe String addition operation is {x+'wow'}")
except ZeroDivisionError:
  print("Zero Division Error!")
except TypeError:
  print("Type Error!")
except ValueError:
  print("Value Error!")
```
