#Program 2 :

Accept angle in degrees from user between 0-90

Print Sin Cos Tan Cosec Cot Sec Values



```python
from types import CodeType
import math
angle = int(input("Enter angle in degrees between 0-90: "))
sin = round(math.sin(math.radians(angle)),2)
cos = round(math.cos(math.radians(angle)),2)
tan = round(sin/cos,2)
cot = round(1/tan,2)
sec = round(1/cos,2)
cosec = round(1/sin,2)
if angle >= 0 and angle <= 90:
  print(f"Sin({angle}) : {sin}\nCos({angle}) : {cos}\nTan({angle}) : {tan}\nCot({angle}) : {cot}\nSec({angle}) : {sec}\nCosec({angle}) : {cosec}")
else:
  print("Invalid Input!")
```

#Program 3 :

Input a number which is divisible by 3 or 5




```python
n = int(input("Enter a number: "))
check = lambda n:n%3==0 and n%5==0
check(n)
```

#Program 4:
